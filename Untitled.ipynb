The following step is loading the two relevant csv files. <br>

```python
import seaborn as sns #this is the plotting library I'll be using 
import pandas as pd #"as pd" means that we can use the abbreviation in commands
import matplotlib.pyplot as plt

df1 = pd.read_csv('survey.csv')
df1.head()
```

```python
df2 = pd.read_csv('steps.csv', sep = ';')
df2.head()
```

Now that we have our two datasets, we want to combine them. <br>
Merging the datasets provides a nice and complete file. After identifying 'id' as a common ground between the two datasets, we use 'id' to merge them.

```python
df = pd.merge(df1, df2, on = 'id')
df.head()
```

Next step is to clean our data. <br>
We cut the unrealistic values from the dataset. Weight above 180 and below 30 is considered unrealistic (generally). We assign these values with a NaN. The NaN allows us to drop these values in their entirity in the next steps.

```python
df['weight'] = df['weight'].apply(pd.to_numeric, errors='coerce')

#To remove the values above 5 we need to create a function that we then apply to the entire column
def above_180(x):
    if(x > 180.0): 
        return float('NaN')
    else: 
        return x
def below_30(x):
    if(x < 30):
        return float('NaN')
    else:
        return x
    
df['weight'] = df['weight'].apply(above_180)
df['weight'] = df['weight'].apply(below_30)

```

A linear regression model is used to predict a certain relationship between a number of variables. <br>
In our case, we have a single regression model, only depicting one dependent variable. <br>
Although I am not sure which one is exactly the dependent variable, the graph seems to indicate that the the more extreme weights/lengths have a higher percentage of variance. <br><br>

While the height distribution (as seen on the right side graph) follows the more classic bell curve/normal distribution. The weight distribution is left skewed, with the majority of participants being around the 75 kg mark.

```python
sns.set_theme(style="darkgrid")

g = sns.jointplot(x="weight", y="height", data=df,
                  kind="reg", truncate=False,
                  xlim=(0, 200), ylim=(140, 200),
                  color="m", height=7)
plt.title('The relation between heigh and weight in France')
plt.axvline(mean_weight, 0, 100, color='blue', label='mean weight')
plt.axhline(mean_height, 0, 100, color='green', label='mean height')
plt.legend() #This adds a legend. It works automatically because I've set the labels in the previous lines
plt.show()
```
